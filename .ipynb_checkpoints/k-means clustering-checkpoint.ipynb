# This notebook provides a na√Øve implementation of the k-means clustering algorithm

We first define a kmeans function that takes in data x, number of centers k, and number of iterations no_of_iterations
It returns the locations of most probable k centroids, representing k clusters in the data x

```python
import numpy as np
from scipy.spatial.distance import cdist 
 
def kmeans(x,k, no_of_iterations):
    #Randomly choose centroids
    idx = np.random.choice(len(x), k, replace=False) 
    centroids = x[idx, :]
     
    #Calculate the distance between centroids and all the data points
    distances = cdist(x, centroids ,'euclidean') 
     
    #Save locations of centroids with the minimum distance
    points = np.array([np.argmin(i) for i in distances]) 
     
    #Repeat the above steps for a defined number of iterations
    for _ in range(no_of_iterations): 
        centroids = []
        for idx in range(k):
            #Update centroid locations by taking mean of corresponding cluster
            temp_cent = x[points==idx].mean(axis=0) 
            centroids.append(temp_cent)
 
        centroids = np.vstack(centroids) #Updated centroids 
         
        distances = cdist(x, centroids ,'euclidean')
        points = np.array([np.argmin(i) for i in distances])
         
    return points 
```

To test our algorithm, we load a toy dataset in sklearn and use the first two principal components as our dataset

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

#Load Data
data = load_digits().data
pca = PCA(2)
  
#Transform the data
df = pca.fit_transform(data)
 
#Applying our k-means algorithm
label = kmeans(df,10,1000)
 
#Visualize the results
u_labels = np.unique(label)
for i in u_labels:
    plt.scatter(df[label == i , 0] , df[label == i , 1] , label = i)
plt.legend()
plt.show()
```
